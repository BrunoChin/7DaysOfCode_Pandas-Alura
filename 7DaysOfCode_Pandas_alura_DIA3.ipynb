## Dia 2 - Manipulação dos dados

Criar uma nova coluna com os valores da localização, para refletir a respectiva classe geral na CDU.

- 000 a 099: Generalidades. Ciência e conhecimento.
- 100 a 199: Filosofia e psicologia.
- 200 a 299: Religião.
- 300 a 399: Ciências sociais.
- 400 a 499: Classe vaga. Provisoriamente não ocupada.
- 500 a 599: Matemática e ciências naturais.
- 600 a 699: Ciências aplicadas.
- 700 a 799: Belas artes.
- 800 a 899: Linguagem. Língua. Linguística.
- 900 a 999: Geografia. Biografia. História."

```python
CDU = []

for loc in dados_emprestimos['localizacao']:
  if loc < 100:
    CDU.append("Generalidade")
  elif loc < 200:
    CDU.append("Filosofia e psicologia")
  elif loc < 300:
    CDU.append("Religião")
  elif loc < 400:
    CDU.append("Ciências Sociais")
  elif loc < 500:
    CDU.append("Classe vaga")
  elif loc < 600:
    CDU.append("Matemática e ciências naturais")
  elif loc < 700:
    CDU.append("Ciências aplicadas")
  elif loc < 800:
    CDU.append("Belas Artes")
  elif loc < 900:
    CDU.append("Linguagem")
  else:
    CDU.append("Geografia, Biografia, Historia")

dados_emprestimos['CDU'] = CDU

dados_emprestimos
```

## Dia 3 - Inciando a analize dos dados

- Quantidade de emprestimos de exemplares por ano
- Periosdo de menor demandas

```python
import plotly.express as px
```

```python
dados_emprestimos['data_emprestimo'] = pd.to_datetime(dados_emprestimos['data_emprestimo'])

dados_emprestimos.info()
```

```python
dados_emprestimos.head()
```

```python
dados_ano = (dados_emprestimos['data_emprestimo'].value_counts()).reset_index()
dados_ano.columns = ['data', 'quantidade']
dados_ano
```

```python
dados_ano.loc[(dados_ano['data'].dt.year == 2014)]
```

```python
total_ano = dados_ano.groupby(dados_ano['data'].dt.year).sum()

total_ano
```

```python
fig = px.line(
    total_ano,
    y='quantidade',
    markers=True,
    title='Emprestimos Por ano',
    labels={'quantidade':'Quantidade', 'data':'Ano'}
)

fig.show()
```

A quantidade de emprestimos deve um queda significativa a partir do ano de 2018. Essa queda fica ainda mais evidente no ano de 2020, mesmo com apensa um semestre de dados, ele apresenta menos de 20% da quantidade de emprestimos em compraração com o ano anterior.

```python
total_mes = pd.DataFrame(dados_ano.groupby(dados_ano['data'].dt.month).sum()).reset_index()
total_mes.columns=['Mes', 'Quantidade']

total_mes
```

```python
fig = px.line(
    total_mes,
    x='Mes',
    y='Quantidade',
    markers=True,
    title='Emprestimos Por Mes'
)

fig.show()
```

```python
dados_mes_ano = dados_ano['quantidade']

dados_mes_ano
```

```python
meses = dados_ano['data'].dt.strftime('%m')
anos = dados_ano['data'].dt.strftime('%Y')
```

```python
dados_mes_ano = pd.concat([dados_mes_ano, meses, anos], axis=1)
```

```python
dados_mes_ano.columns=['Quantidade', 'Mes', 'Ano']

dados_mes_ano
```

```python
total_mes_ano = pd.DataFrame(dados_mes_ano.groupby(['Ano', 'Mes']).sum()).reset_index()
```

```python
total_mes_ano
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure()

for ano in total_mes_ano['Ano'].unique():
  dados = total_mes_ano.loc[(total_mes_ano['Ano'] == ano)]
  fig.add_trace(go.Scatter(x=dados['Mes'], y=dados['Quantidade'], name=ano))

fig.show()
```
